## Q 5. Find the net revenue generated by the E-Commerce company over all orders placed. (6 marks)

### Ans 5 Net revenue is the revenue a company earns in a given period after any purchaser discounts or allowances are factored.
### We need to place the above given conditions of discounted price in the def.
### we need to returnt the value and caluculate the net revenue

```python
#placing the given discount conditions in the def.
def price_of_ECom_Data (temp_df):
    if temp_df.orders > 600:
        ECom_Price = temp_df.discounted_price *0.25
    elif temp_df.orders > 350 and temp_df.orders <= 600 :
       ECom_Price = temp_df.discounted_price *0.15
    elif temp_df.orders > 100 and temp_df.orders <= 350 :
         ECom_Price  = temp_df.discounted_price *0.10
    else:
          ECom_Price  = temp_df.discounted_price *0.05
    return  ECom_Price 
```

```python
#Caluculating the net revenue.
NET_Revenue = ECom_Data.groupby(['Brand']).agg({'Customer_uniq_id':'count','discounted_price':'sum'}).reset_index()\
                                .rename({'Customer_uniq_id':'orders'},axis=1)
NET_Revenue['ECom_Price'] = NET_Revenue.apply(price_of_ECom_Data, axis = 1)
print('Total revenue of E-Commerce company', round(NET_Revenue['ECom_Price'].sum(),2))
```
